```python
from constants import *
import pandas as pd
from utils import pandas_utils
import os
```

```python
rb_df = pd.read_csv(REWARD_BENCH_PATH)

base_suffix = 'orig'
mod_suffix = 'vpAAL'
prompt_strategy = 'w_prompt'
df = rb_df
results = {}

for model in MODELS_DICT.keys():
    # get relevant score col names
    base_chos_score_col = pandas_utils.get_score_col_name(model, prompt_strategy, base_suffix, 'chosen')
    base_rej_score_col = pandas_utils.get_score_col_name(model, prompt_strategy, base_suffix, 'rejected')
    mod_chos_score_col = pandas_utils.get_score_col_name(model, prompt_strategy, mod_suffix, 'chosen')
    mod_rej_score_col = pandas_utils.get_score_col_name(model, prompt_strategy, mod_suffix, 'rejected')

    # get correctly scored items for base and mod texts
    base_correct = df[base_chos_score_col] > df[base_rej_score_col]
    mod_correct = df[mod_chos_score_col] > df[mod_rej_score_col]

    base_y_mod_n = (base_correct == True) & (mod_correct == False)
    base_y_mod_n_df = df[base_y_mod_n]

    base_pred_score_col = f'orig$w_prompt${model}$chosen'
    mod_not_pred_score_col = f'vpAAL$w_prompt${model}$chosen'
    base_y_mod_n_df['diff'] = base_y_mod_n_df[base_pred_score_col] - base_y_mod_n_df[mod_not_pred_score_col]

    base_y_mod_n_df.sort_values(by=['diff'], ascending=False, inplace=True)
    cols_to_keep = ['index', 'txt_prompt_orig', 'txt_chos_orig', 'txt_rej_orig', 'txt_prompt_vpAAL', 'txt_chos_vpAAL', 'txt_rej_vpAAL', 'diff']
    base_y_mod_n_df = base_y_mod_n_df[cols_to_keep]

    print(model, len(base_y_mod_n_df))
    path = f'{ERROR_ANALYSIS_DIR}/base_y_mod_n${model}.csv'
    os.makedirs(os.path.dirname(path), exist_ok=True)
    pandas_utils.save_df(base_y_mod_n_df, path)
```
